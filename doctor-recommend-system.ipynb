```python
import pandas as pd
import numpy as np
import os
import joblib
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
try:
    df = pd.read_csv('/kaggle/input/doctors-data/doctors_dataset.csv')
    print("Successfully loaded dataset")
    
    df['Specialist'] = df['Specialist'].fillna('')
    df['Tags'] = df['Tags'].fillna('')
    
    df['combined_text'] = df['Specialist'].astype(str) + " " + df['Tags'].astype(str)
    
    tfidf = TfidfVectorizer(stop_words='english')
    
    tfidf_matrix = tfidf.fit_transform(df['combined_text'])
    
    joblib.dump(tfidf, 'tfidf_vectorizer.pkl')
    joblib.dump(tfidf_matrix, 'tfidf_matrix.pkl')
    joblib.dump(df, 'doctors_df.pkl')
    
    print("\nModel trained and .pkl files generated successfully!")

except FileNotFoundError:
    print("Error: File not found. Please double-check the '/kaggle/input/...' path.")
```
