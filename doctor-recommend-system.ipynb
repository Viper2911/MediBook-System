```python
import pandas as pd
import numpy as np
import os
import joblib
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
try:
    df=pd.read_csv('/kaggle/input/doctors-data/doctors_dataset.csv')
    print("Successfully loaded dataset")
except FileNotFoundError:
    print("Error, File not found")
```

```python
df['combined_text']=df['Specialist'].astype(str)+" "+df['Tags'].astype(str)
```

```python
tfidf=TfidfVectorizer(stop_words='english')
tfidf_matrix=tfidf.fit_transform(df['combined_text'])
```

```python
joblib.dump(tfidf, 'tfidf_vectorizer.pkl')
joblib.dump(tfidf_matrix, 'tfidf_matrix.pkl')
joblib.dump(df, 'doctors_df.pkl')

print("\n Model trained successfully on all 30 doctors!")
```
